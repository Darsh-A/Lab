```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib widget
```

```python
data = np.loadtxt("data.txt",delimiter=",",dtype='float',skiprows=1)
data2 = np.loadtxt("data2.txt",delimiter=",",dtype='float',skiprows=1)
```

```python

def plot(x,y):
    m1,c1 = np.polyfit(x,y,1)
    xFit = np.arange(0,5,0.1)
    plt.clf()
    plt.plot(x,y,'.')
    plt.plot(xFit,(m1*xFit)+c1)
    plt.show()
    return (m1,c1)
```

```python
val1 = plot(data[:,0],data[:,2])
plt.title("Uc vs Q (Air)")
print(val1)
e1 = (0.002/0.053)/val1[0]
print(e1)
```

```python
val2 = plot(data2[:,0],data2[:,1])
print(val2)
e2 = (0.0115/0.0531)/val2[0]
print(e2)
```
