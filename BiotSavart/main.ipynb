```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import json
%matplotlib widget

Colors = json.load(open('../config.json'))
```

```python
# Loading Data

# Note:
# Var_Current follows a current,mang field pattern
# Var_Dist follows a dist,mang field pattern

# For coil 120
Var_Current_120 = np.loadtxt('data/Var_Current_120.txt',delimiter=',',dtype='float')
Var_Dist_120 = np.loadtxt('data/Var_Dist_120.txt',delimiter=',',dtype='float')

# For coil 80
Var_Current_80 = np.loadtxt('data/Var_Current_80.txt',delimiter=',',dtype='float')
Var_Dist_80 = np.loadtxt('data/Var_Dist_80.txt',delimiter=',',dtype='float')

# For coil 40
Var_Current_40 = np.loadtxt('data/Var_Current_40.txt',delimiter=',',dtype='float')
Var_Dist_40 = np.loadtxt('data/Var_Dist_40.txt',delimiter=',',dtype='float')
```

```python
# Plotting Current vs Magnetic Field

# For 120
Current_120 = Var_Current_120[:,0]
Magn_120 = Var_Current_120[:,1]

# For 80
Current_80 = Var_Current_80[:,0]
Magn_80 = Var_Current_80[:,1]

# For 40
Current_40 = Var_Current_40[:,0]
Magn_40 = Var_Current_40[:,1]

plt.clf()
plt.scatter(Current_120,Magn_120,color='#688bfc')
plt.scatter(Current_80,Magn_80,color='#b468fc')
plt.scatter(Current_40,Magn_40,color='#fa436b')
plt.plot(Current_120,Magn_120,color='#688bfc', label='diameter: 120')
plt.plot(Current_80,Magn_80,color='#b468fc', label='diameter: 80')
plt.plot(Current_40,Magn_40,color='#fa436b', label='diameter:40')
plt.xlabel("Current")
plt.ylabel("Magnetic Field")
plt.legend()
plt.show()

```

```python

def posplot(Dist,MagnDist):
    NewDist = []
    for i in Dist:
        NewDist.append(43.9-i)

    return NewDist,MagnDist

def negplot(Dist,MagnDist):
    NewDist = []
    for i in Dist:
        NewDist.append(i-43.9)

    return NewDist,MagnDist

plt.clf()
plt.scatter(posplot(Var_Dist_120[:,0],Var_Dist_120[:,1])[0],posplot(Var_Dist_120[:,0],Var_Dist_120[:,1])[1],color='#688bfc',marker='.')
plt.scatter(negplot(Var_Dist_120[:,0],Var_Dist_120[:,1])[0],negplot(Var_Dist_120[:,0],Var_Dist_120[:,1])[1],color='#688bfc',marker='.')

plt.scatter(posplot(Var_Dist_80[:,0],Var_Dist_80[:,1])[0],posplot(Var_Dist_80[:,0],Var_Dist_80[:,1])[1],color='#b468fc',marker='.')
plt.scatter(negplot(Var_Dist_80[:,0],Var_Dist_80[:,1])[0],negplot(Var_Dist_80[:,0],Var_Dist_80[:,1])[1],color='#b468fc',marker='.')

plt.scatter(posplot(Var_Dist_40[:,0],Var_Dist_40[:,1])[0],posplot(Var_Dist_40[:,0],Var_Dist_40[:,1])[1],color='#fa436b',marker='.')
plt.scatter(negplot(Var_Dist_40[:,0],Var_Dist_40[:,1])[0],negplot(Var_Dist_40[:,0],Var_Dist_40[:,1])[1],color='#fa436b',marker='.')

plt.plot(posplot(Var_Dist_120[:,0],Var_Dist_120[:,1])[0],posplot(Var_Dist_120[:,0],Var_Dist_120[:,1])[1],color='#688bfc', label='diameter: 120')
plt.plot(negplot(Var_Dist_120[:,0],Var_Dist_120[:,1])[0],negplot(Var_Dist_120[:,0],Var_Dist_120[:,1])[1],color='#688bfc')

plt.plot(posplot(Var_Dist_80[:,0],Var_Dist_80[:,1])[0],posplot(Var_Dist_80[:,0],Var_Dist_80[:,1])[1],color='#b468fc', label='diameter: 80')
plt.plot(negplot(Var_Dist_80[:,0],Var_Dist_80[:,1])[0],negplot(Var_Dist_80[:,0],Var_Dist_80[:,1])[1],color='#b468fc')

plt.plot(posplot(Var_Dist_40[:,0],Var_Dist_40[:,1])[0],posplot(Var_Dist_40[:,0],Var_Dist_40[:,1])[1],color='#fa436b', label='diameter: 40')
plt.plot(negplot(Var_Dist_40[:,0],Var_Dist_40[:,1])[0],negplot(Var_Dist_40[:,0],Var_Dist_40[:,1])[1],color='#fa436b')

plt.xlabel("Distance")
plt.ylabel("Magnetic Field")
plt.legend()

plt.show()
    
```
