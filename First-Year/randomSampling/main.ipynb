```python
# Loading Data

data = np.loadtxt("data500")
data2 = np.loadtxt("data2-500")

data100 = np.loadtxt("data100")
data2100 = np.loadtxt("data2-100")

data.shape
```

```python
# Histogram Plots with different Bin Size

def histPlot(data,bins):
    hist_data, bin_edges = np.histogram(data,bins=bins)
    plt.clf()
    for i in range(bins):
        for j in range(1,hist_data[i]+1):
            plt.plot(bin_edges[i],j,'rx')
    plt.title("Histogram of 100 samples (binSize = 0.5)")
    plt.xlabel('Voltage')
    plt.ylabel('Frequency')
    plt.minorticks_on()
    plt.grid()
    plt.show()


histPlot(data,9)
```

```python
# Normalised Histogram

def probDist(Vo,V):
    return 1/(np.pi*( np.sqrt( (Vo**2) - (V**2) ) ))

plt.clf()
plt.hist(data2,bins=18,density=True,histtype='step',color="#5fd9ac")
Xdata = np.linspace(-9.1,9.1,500)
plt.plot(Xdata,probDist(9.1,Xdata),color="#eb3459")
plt.ylabel("nv")
plt.xlabel("V")
plt.ylim(0,0.25)
plt.title("Probability Distribution")
plt.minorticks_on()
plt.grid()
plt.show()
```

```python
def Nv(Vo,V,N):
    return (N/np.pi)*(np.arcsin(V/Vo))
def volFun(Vo,N,Nv):
    return Vo*np.sin(np.pi*Nv/N)

volrange = np.arange(0,9.1,0.5)
Xval = Nv(9.1,volrange,500)
Yval = volFun(9.1,500,volrange)
Yval = volrange

plt.clf()
plt.title("Nv vs V (Recovered waveform)")
plt.plot(Xval,Yval,'ro',mfc='none')
plt.xlabel("Nv")
plt.ylabel("V=Vo*sin(pi*Nv/N)")
plt.minorticks_on()
plt.grid()
plt.show()
```
